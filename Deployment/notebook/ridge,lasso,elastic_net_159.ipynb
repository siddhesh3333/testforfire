```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df.columns
```

## droping month , day , year

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.isnull().value_counts()
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

#### Encoding

```python
df['Classes'] = np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

### Independent and depenedent feature

```python
X =df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
y.head()
```

```python
####  Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

#### Feature Selection based on  correlation

```python
X_train.corr()
```

```python
#### feature Selection  based on correlation
X_train.corr()
```

```python
#### checking the  Multicollinearity
plt.figure(figsize=(5,5))
corr = X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def  correaltion(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in  range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i ,j]) > threshold:
                colname =corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

####  Threshold Domain Expertise


```python
corr_features = correaltion(X_train,0.85)
```

```python
corr_features
```

##### Dropping the features  when correlation is more than 0.85

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis =1,inplace=True)
X_train.shape,X_test.shape
```

#### Feature scaling OR stabdardization 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_scaled =scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

#### Box plot  to understand the effect of Standard Scaler

```python
plt.subplots(figsize=(7,7))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train before Scaling")
plt.subplot(1,2,2) 
sns.boxplot(data=X_train_scaled)   

plt.title(" X_train  after scaling")
```

#### Linear Regression model

```python
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg = LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred = linreg.predict(X_test_scaled)
mae =mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute error", mae)
print("R2 Score",score)
plt.scatter(y_test,y_pred)
```

## Lasso  Regression
### SIDDHESH DHANDE

```python
from sklearn.linear_model  import  Lasso
from sklearn.metrics import  mean_absolute_error ,r2_score
lasso = Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred  = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("mean Absolute error",mae)

print("R2 Score",score)
plt.scatter(y_test,y_pred)
```
